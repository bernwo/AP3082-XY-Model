## Import the required libraries.


```python
import numpy as np
from Metropolis_functions import Metropolis, creategif
from Observables_functions import get_energy_per_spin_per_lattice
import matplotlib.pyplot as plt
```

## Initialise the required variables

```python
J = 1 # coupling constant, where it must be > 0 for the system to exhibit ferromagnetism
L = 32 # dimension of the square lattice

tau = 0 # relaxation time a.k.a equilibrating time
tau_extra = 200000
nrelax = 2 # total number of times to let the system equilibrate
nframes = 20 # total number of frames in the .gif over the whole simulation (not needed since we are not generating the animated .gif in this notebook)
N = nrelax * (tau+tau_extra) # total number of iterations in the Metropolis algorithm
plot_at_Nth_index = np.linspace(0, N - 1, nframes).astype(np.int32) # (not needed since we are not generating the animated .gif in this notebook)

# lattice = np.zeros((L, L))
lattice = np.random.uniform(low=-np.pi, high=np.pi, size=(L,L))

Tc = 0.892937 * J  # http://www.lps.ens.fr/~krauth/images/7/72/Stage_Mayer_Johannes_2015.pdf in units of kB
T_init = 0.00001*Tc
T_final = 3*Tc
TTc_array = np.linspace(T_init, T_final, nrelax)

save_for_plot=True
```

## Begin the Metropolis time-marching on the lattice

```python
ave_M2, ave_E, ave_E2, lattices_plot, index_history = Metropolis(
    J=J,
    L=L,
    relaxation_time=tau,
    extra_time=tau_extra,
    lattice=lattice,
    T_init=T_init,
    T_final=T_final,
    T_n=nrelax,
    plot_at_Nth_index=plot_at_Nth_index,
    save_for_plot=save_for_plot)
```

```python
make_gif = True
if make_gif:
    creategif(J=J,
            L=L,
            T=TTc_array,
            Tc=Tc,
            ave_M2=np.mean(ave_M2,axis=1),
            Cv=(np.mean(ave_E2,axis=1)-np.mean(ave_E,axis=1)**2)/((L**2*TTc_array**2)),
            plot_at_Nth_index=plot_at_Nth_index,
            lattices_plot=lattices_plot,
            index_history=index_history,
            filename=f"test_images/Metropolis_phase_energy_observables_L{L}_tau{tau}_tauextra{tau_extra}_nrelax{nrelax}_Tinit{np.round(T_init/Tc,3)}_Tfinal{np.round(T_final/Tc,3)}_nframes{nframes}.gif",
            plot_mode='phase_and_energy_spaces_arrows')
```
